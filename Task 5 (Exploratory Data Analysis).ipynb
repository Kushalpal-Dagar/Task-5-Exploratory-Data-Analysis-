```python
sns.pairplot(df)
plt.suptitle("Pairplot of Numeric Variables", y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Observation: Age and Fare show visible clusters when compared with Survived. SibSp & Parch are positively related.
```

```python
# ---------------- C1. Histograms ----------------
df.hist(bins=20, figsize=(12, 8), grid=False, edgecolor='black')
plt.suptitle("Histograms of Numeric Features", y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Observation: Fare and Age are right-skewed. Most passengers are in 3rd class and young adults.
```

```python
# ---------------- C2. Boxplots ----------------
plt.figure(figsize=(12, 8))
for i, column in enumerate(df.columns):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(y=df[column])
    plt.title(f"Boxplot of {column}")
plt.tight_layout()
plt.show()
```

```python
# Observation: Outliers seen in Fare and Age. Age includes some passengers over 70.
```

```python
# ---------------- C3. Scatterplot ----------------
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='Set1')
plt.title("Scatterplot: Age vs Fare colored by Survival")
plt.show()
```

```python
# Observation: Higher Fare and younger passengers had a better chance of survival.
```

```python
# ---------------- Optional: Categorical Plots ----------------
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival Count by Sex')
plt.show()
```

```python
# Observation: Females had a significantly higher survival rate than males.
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival Count by Passenger Class')
plt.show()
```

```python
# Observation: 1st class had highest survival rate; 3rd class had the lowest.
```
